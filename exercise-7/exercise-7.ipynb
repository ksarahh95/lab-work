```python
in_file = open("/home/eeb177-student/Desktop/eeb-177/class-assignments-2/classwork-21-feb/cocoli.txt")
in_file.readline() # getting rid of header
sizes_94 = {}
sizes_97 = {}
sizes_98 = {}

for line in in_file:
    items = line.split()
    tree_id = items[0]
    size_in_94 = float(items[4])
    size_in_97 = float(items[5])
    size_in_98 = float(items[6])
    sizes_94[tree_id] = size_in_94
```
